#### Importación de Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importación de datos



```python
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Visualizando informaciones de las columnas
tienda.info()
```

## 1. Análisis de facturación

```python
# Ingresos totales de cada tienda

ing_t1 = tienda['Precio'].sum()
ing_t2 = tienda2['Precio'].sum()
ing_t3 = tienda3['Precio'].sum()
ing_t4 = tienda4['Precio'].sum()
```

```python
print(f"Ingresos totales de la tienda 1: {ing_t1:,}")
print(f"Ingresos totales de la tienda 2: {ing_t2:,}")
print(f"Ingresos totales de la tienda 3: {ing_t3:,}")
print(f"Ingresos totales de la tienda 4: {ing_t4:,}")
```

## 2. Ventas por categoría

## 3. Calificación promedio de la tienda


```python
# Promedio de calificación de cada tienda

prom_calif_t1 = tienda['Calificación'].mean()
prom_calif_t2 = tienda2['Calificación'].mean()
prom_calif_t3 = tienda3['Calificación'].mean()
prom_calif_t4 = tienda4['Calificación'].mean()
```

## Informe de Desempeño Operativo y Estrategia de Mercado

Este análisis examina el rendimiento comparativo de nuestras cuatro unidades de negocio, evaluando no solo el músculo financiero de cada tienda, sino también la percepción del cliente y la eficiencia logística. Los datos nos permiten trazar una hoja de ruta clara para la toma de decisiones estratégicas.

---

### Análisis de Situación y Diagnóstico

Al profundizar en los números, observamos un fenómeno interesante. La **Tienda 1** se consolida como la auténtica locomotora de la organización en términos de facturación. Es la unidad que mueve el mayor volumen de productos y genera el flujo de caja más robusto. Sin embargo, este liderazgo tiene un precio: es la tienda con la calificación de satisfacción más baja y los costes de envío más elevados, lo que sugiere que su éxito actual podría verse amenazado si no se optimiza la logística y la atención al cliente a largo plazo.

En el otro extremo del espectro, la **Tienda 3** ha logrado encontrar un "punto dulce" operativo. Aunque su volumen de ventas es un % inferior al de la Tienda 1, lidera la satisfacción de los compradores con una calificación de 4.05. Esto nos indica una operación más saneada y un cliente más fiel, apoyado en costes de envío equilibrados.

Por último, la **Tienda 4** presenta el panorama más complejo. A pesar de contar con la ventaja competitiva de ofrecer el envío más económico del grupo, no logra capitalizar este beneficio en volumen de ventas ni en una calificación superior, quedando rezagada como el eslabón más débil de la cadena.

---

### Evaluación Estratégica: ¿Qué unidad se podría vender?

La decisión de vender o no invertir en una tienda no depende de un solo número, sino de la estrategia corporativa que se quiera seguir. Hemos identificado dos escenarios posibles:

#### Escenario A: Maximización de valor (Vender la "Joya de la Corona")
Si el objetivo es poner en el mercado un activo de alto valor para obtener liquidez inmediata, la **Tienda 1** es la candidata ideal. Cualquier inversionista se sentirá atraído por su capacidad de generación de ingresos y su dominancia en la rotación de inventario. Es un negocio probado que, a pesar de sus áreas de mejora en costes y servicio, representa una inversión muy atractiva por su escala.

#### Escenario B: Optimización del Portafolio (Eliminar la ineficiencia)
Si lo que buscamos es sanear la estructura del grupo y eliminar las unidades que lastran el crecimiento promedio, la decisión lógica es desprenderse de la **Tienda 4**. Sus bajos ingresos y su rendimiento por debajo de la media la convierten en una unidad que requiere demasiado esfuerzo para resultados mediocres. El no invertir aquí, permitiría concentrar recursos en potenciar la eficiencia de las otras tres tiendas.

---

### Conclusión y Recomendación Final

En definitiva, estamos ante una elección de perspectiva. La **Tienda 1** es nuestro activo más valioso para un comprador externo, mientras que la **Tienda 4** es nuestro activo menos eficiente internamente. 

Basándonos en la estructura de ventas, recomendamos que cualquier movimiento de venta se respalde con los gráficos de volumen de productos, demostrando que incluso en la unidad menos rentable, existe una base sólida de demanda que un nuevo dueño podría explotar bajo una gestión diferente.
